```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

```python
students_data=pd.read_csv('students_data.csv')
students_data
```

```python
plt.style.use('seaborn-whitegrid')
fig, (ax0,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                              figsize=(30,15));

ax0.plot(students_data['Student Name'],
         students_data['First C.A'],)

ax0.set_title(label = 'Distribution of Student CA and the average score of the Class', fontsize= 25)

for tick in ax0.get_xticklabels():
    tick.set_fontsize(14)
for tick in ax0.get_yticklabels():
    tick.set_fontsize(16)

ax0.set_ylabel('C.A score' , fontsize=25 )

ax0.axhline (y = students_data['First C.A'].mean(),
               linestyle = '--');

ax1.plot(students_data['Student Name'],
        students_data['Lab test'])

ax1.set_title(label='Distribution of Students Lab Test and the average score of the class' , fontsize= 25)

for tick in ax1.get_xticklabels():
    tick.set_fontsize(14)
for tick in ax1.get_yticklabels():
    tick.set_fontsize(16)

ax1.set_xlabel('Student Name' , fontsize=20)
ax1.set_ylabel('Lab test' , fontsize= 25)
ax1.axhline(y =students_data['Lab test'].mean(),
            linestyle = '--')

fig.suptitle('Student Data Analysis', fontsize=40 ,fontstyle='oblique');
```
